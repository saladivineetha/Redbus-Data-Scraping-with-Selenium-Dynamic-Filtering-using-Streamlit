# SQL Connection

```python
# Establish connection to MySQL
myconnection = pymysql.connect(host='localhost', user='root', password='root', database='REDBUS_DETAILS')
my_cursor = myconnection.cursor()
my_cursor.execute("CREATE DATABASE IF NOT EXISTS RED_BUS_DETAILS")
```

```python
with myconnection:
     my_cursor = myconnection.cursor()
     my_cursor.execute("CREATE DATABASE IF NOT EXISTS RED_BUS_DETAILS")
     # Table Creation
     my_cursor.execute('''CREATE TABLE IF NOT EXISTS redbus_complete_details(
                  ID INT AUTO_INCREMENT PRIMARY KEY,
                  Bus_name VARCHAR(255) NOT NULL,
                  Bus_type VARCHAR(255) NOT NULL,
                  Start_time VARCHAR(255) NOT NULL,
                  End_time VARCHAR(255) NOT NULL,
                  Total_duration VARCHAR(255) NOT NULL,
                  Price FLOAT NULL,
                  Seats_Available VARCHAR(255) NOT NULL,
                  Ratings Float NULL,
                  Route_link VARCHAR(255) NULL,
                  Route_name VARCHAR(255) NULL
                  )''')
     
     # SQL query to insert data into bus_details table
     insert_query = '''INSERT INTO redbus_complete_details(
                    Bus_name,
                    Bus_type,
                    Start_time,
                    End_time,
                    Total_duration,
                    Price,
                    Seats_Available,
                    Ratings,
                    Route_link,
                    Route_name)
                  VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'''
     data = FinalRedBus_df.values.tolist()

     my_cursor.executemany(insert_query, data)

     myconnection.commit()

print("Values inserted successfully")
```
